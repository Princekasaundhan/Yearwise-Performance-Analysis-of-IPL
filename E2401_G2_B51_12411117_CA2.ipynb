```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data
# Update with your file path
data = pd.read_csv('IPL2024_Data.csv')

# Ensure numeric conversion for columns where aggregation will be applied
numeric_columns = ["Matches_Batted", "Runs_Scored", "Batting_Average", "Batting_Strike_Rate"]
for col in numeric_columns:
    if col in data.columns:
        data[col] = pd.to_numeric(data[col], errors="coerce")

# Player names for comparison
players = ["Jitesh Sharma", "Jonny Bairstow", "Prabhsimran Singh"]

# Filter data for selected players
player_data = data[data["Player_Name"].isin(players)]

# Group data by player and year
player_stats = player_data.groupby(["Player_Name", "Year"]).agg({
    "Matches_Batted": "sum",
    "Runs_Scored": "sum",
    "Batting_Average": "mean",
    "Batting_Strike_Rate": "mean"
}).reset_index()
```

```python
# Plot 1: Bar plot for Runs Scored
plt.figure(figsize=(10, 6))
for player in players:
    stats = player_stats[player_stats["Player_Name"] == player]
    plt.bar(stats["Year"], stats["Runs_Scored"], label=player, alpha=0.7)
plt.title("Runs Scored Comparison")
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Line plot for Batting Average
plt.figure(figsize=(10, 6))
for player in players:
    stats = player_stats[player_stats["Player_Name"] == player]
    plt.plot(stats["Year"], stats["Batting_Average"], label=player, marker="o")
plt.title("Batting Average Comparison")
plt.xlabel("Year")
plt.ylabel("Batting Average")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plot 3: Line plot for Batting Strike Rate
plt.figure(figsize=(10, 6))
for player in players:
    stats = player_stats[player_stats["Player_Name"] == player]
    plt.plot(stats["Year"], stats["Batting_Strike_Rate"], label=player, marker="o")
plt.title("Batting Strike Rate Comparison")
plt.xlabel("Year")
plt.ylabel("Batting Strike Rate")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plot 4: Bar plot for Matches Batted
plt.figure(figsize=(10, 6))
for player in players:
    stats = player_stats[player_stats["Player_Name"] == player]
    plt.bar(stats["Year"], stats["Matches_Batted"], label=player, alpha=0.7)
plt.title("Matches Batted Comparison")
plt.xlabel("Year")
plt.ylabel("Matches Batted")
plt.legend()
plt.tight_layout()
plt.show()
```
